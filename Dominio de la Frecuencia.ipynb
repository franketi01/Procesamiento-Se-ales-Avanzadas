```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300  # NÃºmero de puntos o muestras
tiempo = np.linspace(1, 300, N)  # Vector de tiempo de 1 a 300
```

```python
ciclos = 1
amplitud = 1

# Calcular seno para ciclos=1
seno1 = amplitud * np.sin(ciclos * 2 * np.pi * tiempo / N)

# Graficar para ciclos=1
plt.plot(tiempo, seno1, label='ciclos=1')
```

```python
ciclos = 2
amplitud = 1
seno2 = amplitud * np.sin(ciclos * 2 * np.pi * tiempo / N)

# Graficar para ciclos=2
plt.plot(tiempo, seno2, label='ciclos=2')
```

```python
# Calcular seno para ciclos=10
ciclos = 10
amplitud = 2
seno10 = amplitud * np.sin(ciclos * 2 * np.pi * tiempo / N)

# Graficar para ciclos=10
plt.plot(tiempo, seno10, label='ciclos=10')

# AÃ±adir leyenda y etiquetas de ejes
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Ondas sinusoidales con diferentes ciclos y amplitudes')

# Mostrar el grÃ¡fico
plt.show()
```

```python
ciclos = 20
amplitud = 0.5

# Calcular seno para ciclos=20
seno20 = amplitud * np.sin(ciclos * 2 * np.pi * tiempo / N)

# Graficar para ciclos=20
plt.plot(tiempo, seno20, label='ciclos=20')

# Calcular nivel DC
niveldc = np.ones(N)

# Graficar nivel DC
plt.plot(tiempo, niveldc, label='nivel DC')

# Calcular la suma de todas las ondas
seno121020dc = seno1 + seno2 + seno10 + seno20 + niveldc

# Graficar la suma de todas las ondas
plt.plot(tiempo, seno121020dc, label='suma de todas las ondas')

# AÃ±adir leyenda y etiquetas de ejes
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Ondas sinusoidales con diferentes ciclos, amplitudes y nivel DC')

# Mostrar el grÃ¡fico
plt.show()
```

```python
fft_seno1 = np.fft.fft(seno1)

# Calcular el mÃ³dulo de la FFT y graficar
plt.plot(np.abs(fft_seno1), label='FFT de seno1')
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud')
plt.title('Magnitud de la FFT de seno1')
plt.show()
```
